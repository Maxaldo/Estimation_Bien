```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv("data.csv")


X = data.drop(columns=["Unnamed: 0", "prix_eur" , "prix_m2" , "url" , "latitude" , "longitude"])
y = data["prix_eur"]


mapping_dpe = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7}
X["dpe"] = X["dpe"].map(mapping_dpe)

le_etage = LabelEncoder()
X["etage"] = le_etage.fit_transform(X["etage"])

le_type_bien = LabelEncoder()
X["type_bien"] = le_type_bien.fit_transform(X["type_bien"])

le_adresse = LabelEncoder()
X["adresse"] = le_adresse.fit_transform(X["adresse"])

X = X.fillna(-10000)

from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor(n_estimators=300, learning_rate=0.05,
                                   max_depth=5, random_state=42)

model.fit(X,y)
```

```python
import joblib
joblib.dump(model, 'mon_modele.joblib')
```

```python
joblib.dump(le_etage, 'le_etage.joblib')
joblib.dump(le_type_bien, 'le_type_bien.joblib')
joblib.dump(le_adresse, 'le_adresse.joblib')
joblib.dump(mapping_dpe, 'mapping_dpe.joblib')
```
